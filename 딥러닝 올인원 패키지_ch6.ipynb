## 생성자(__init__) 이해하기

```python
# 생성자, 클래스 인스턴스가 생성될 때 호출됨

class Person:
    def __init__(self):
        print(self, 'is generated')
        self.name = 'Kate'
        self.age = 10
        
p1 = Person()
p2 = Person()

p1.name = 'aaron'
p1.age = 20

print(p1.name, p1.age)
print(p2.name, p2.age)
```

```python
class Person:
    def __init__(self, name, age = 10):
        print(self, 'is generated')
        
        # 관례적으로 같게함.
        self.name = name
        self.age = age
        
p1 = Person('Bob', 30)
p2 = Person('Kate', 20)
p3 = Person('Aaron')

print(p1.name, p1.age)
print(p2.name, p2.age)
print(p3.name, p3.age)
```

##### self  
C++/C#, Java의 this에 해당된다.  
위치는 항상 맨 처음의 parameter이며 관례적으로 self로 사용.
파이썬의 method는 항상 첫번째 인자로 self를 전달  

```python
class Person:
    def __init__(self, name, age):
        print('self: ', self)
        self.name = name
        self.age = age
        
    def sleep(a):
        print('self: ', a)
        print(a.name, '은 잠을 잡니다.')
        
a = Person('Aaron', 20)
b = Person('Bob', 30)

print(a) # 객체가 메모리상에 상주하고 있는 주소
print(b)

a.sleep()
#b.sleep()
```

### method 정의, method type

```python
# 1. 숫자를 하나 증가
# 2. 숫자를 0으로 초기화

class Counter:
    def __init__(self):
        self.num = 0
    
    def increment(self):
        self.num += 1
        
    def reset(self):
        self.num = 0
    
    def print_current_value(self):
        print('현재값 : ', self.num)
        

#Counter : 클래스 레벨
#c1 : 객체 레벨
c1 = Counter()
c1.print_current_value()
c1.increment()
c1.increment()
c1.increment()
c1.print_current_value()

c1.reset()

c1.print_current_value()

# c2 = Counter()
# c2.print_current_value()
```

## Class Inheritance(상속)  
기존에 정의해둔 클래스의 기능을 그대로 물려받을 수 있다.  
기존 클래스에 기능 일부를 추가하거나, 변경하여 새로운 클래스로 정의한다.

```python
class Person: #공통 기능들을 부모 클래스로 하나로 생성
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self, food):
        print('{}은 {}를 먹습니다.'.format(self.name, food))
    
    def sleep(self, minute):
        print('{}은 {}분 동안 잡니다.'.format(self.name, minute))
    
    def work(self, minute):
        print('{}은 {}분 동안 일합니다.'.format(self.name, minute))
    
class Student(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age 
    
    # 위의 예제를 보면, 문제점은 Student는 
    # work가 아닌 study를 하기 때문에 재정의할 필요가 있다.
    # Override
    def work(self, minute):
        print('{}은 {}분 동안 공부합니다.'.format(self.name, minute))       
        
class Employee(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def work(self, minute):
        print('{}은 {}분 동안 업무를 합니다.'.format(self.name, minute))  
bob = Employee('Bob', 25)
bob.eat('BBQ')
bob.sleep(30)
bob.work(60)
```

###  super method  
하위클래스(자식 클래스)에서 부모 클래스를 호출할 때 사용

```python
class Person: #공통 기능들을 부모 클래스로 하나로 생성
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self, food):
        print('{}은 {}를 먹습니다.'.format(self.name, food))
    
    def sleep(self, minute):
        print('{}은 {}분 동안 잡니다.'.format(self.name, minute))
    
    def work(self, minute):
        print('{}은 {}분동안 일할 준비를 합니다.'.format(self.name, minute))
    
class Student(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age 

    def work(self, minute):
        super().work(minute)
        print('{}은 {}분 동안 공부합니다.'.format(self.name, minute))       
        
class Employee(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    def work(self, minute):
        super().work(minute)
        print('{}은 {}분 동안 업무를 합니다.'.format(self.name, minute))  
        
bob = Employee('Bob', 25)
bob.eat('BBQ')
bob.sleep(30)
bob.work(60)
```

```python
# Point
# 2차원 좌표평면 각 점(x, y)
# 연산
# 두점의 덧셈, 뺄셈
# 한점과 숫자의 곱셈 (1, 2) * 3 = (3, 6)
# 그 점의 길이 (0, 0) 부터의 거리
# x, y 값 가져오기

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point(new_x, new_y)
    
    def __sub__(self, pt):
        new_x = self.x - pt.x
        new_y = self.y - pt.y
        return Point(new_x, new_y)
    
    def __mul__(self, factor):
        return Point(self.x * factor, self.y * factor)
    
    def __getitem__(self, index):
        if index == 0:
            return self.x
        elif index == 1:
            return self.y
        else:
            return -1
    
    def __len__(self):
        return self.x ** 2 + self.y ** 2
    
    def __str__(self):
        return '({}, {})'.format(self.x, self.y)

p1 = Point(3, 4)
p2 = Point(2, 7)

p3 = p1 + p2
p4 = p1 - p2
p5 = p1 * 3

print(len('hello'))
print(len(p1))

print(p1[0])
print(p1[1])

print(p1)
print(p2)
print(p3)
print(p4)
print(p5)
```

```python
import math

# 복소수 클래스를 정의
# 덧셈, 뺄셈, 곱셈 연산자 지원
# 길이(크기) 지원
# 복소수 출력 '1 +4j'와 같이 표현
# 비교 연산 ==, != 지원
# =, <=, <, > 연산 지원
# 절대값 지원

class ComplexNumber:
    def __init__(self, real, img):
        self.real = real
        self.img = img
        
    def __add__(self, cn):
        return ComplexNumber(self.real + cn.real, self.img + cn.img)
    
    def __sub__(self, cn):
        return ComplexNumber(self.real - cn.real, self.img - cn.img)
    
    def __mul__(self, x):
        if type(x) == int:
            return ComplexNumber(self.real * x, self.img * x)
        elif type(x) == ComplexNumber:
            return ComplexNumber(self.real * x.real - self.img * x.img, self.real * x.img + self.img * x.real)
            # (a + bj) * (c + dj) = (ac - bd) + (ad + bc)j
    
    def __str__(self):
        if self.img >= 0:
            return '{} + {}j'.format(self.real, self.img)
        else:
            return '{} - {}j'.format(self.real, abs(self.img))
    
    def __eq__(self, cn):
        return self.real == cn.real and self.img == cn.img
    
    def __ne__(self, cn):
        return not(self.real == cn.real and self.img == cn.img)
    
    def __abs__(self):
        return math.sqrt(self.real ** 2 + self.img ** 2)
    
    
    
a = ComplexNumber(1, 2)
b = ComplexNumber(3, 5)
print(a)

print(a * b)
    
abs(a)
```
