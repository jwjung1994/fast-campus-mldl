```python
# a = [22, 1, 3 , 4, 7, 98, 21, 55, 87, 99, 19, 20, 45]에서 최대값과 최소값을 찾으시오
# sorted, sort 사용 X
a = [22, 1, 3 , 4, 7, 98, 21, 55, 87, 99, 19, 20, 45]
max_value = a[0]
min_value = a[0]
for _, item in enumerate(a):
    if _ > 0: # 1부터
        if(max_value < item):
            max_value = item
        if(min_value > item):
            min_value = item
            
print(max_value, min_value)
```

## multiple return (복수 값 변환)  
tuple반환을 하여 복수개의 값 리턴 가능

```python
def add_mul(x, y):
    s = x + y
    m = x * y
    
    return s, m # 사실 tuple을 반환시키는 개념
```

```python
c = add_mul(20, 3)
print(type(c))
print(c)
```

```python
c1, c2 = add_mul(20, 3)


print(type(c1))
print(type(c2))
print(c1)
print(c2)
```

## variable length argument (가변길이 인자)  
전달되는 파라미터의 개수가 고정적이지 않은 경우 사용  
*args : 파라미터를 튜플의 형태로 전달  
***kwargs : 파라미터를 딕셔너리 형태로 전달

```python
def test(*x):
    print(type(x))
    
test(10, 20)
```

```python
#가변길이 파라미터를 사용할 땐, 관례적으로 args라는 이름으로 사용한다.
def test(*args): # arguments
    print(type(args))
    for item in args:
        print(item, end = ' ')

        
test(10, 20, 30 ,40, 50)
```

```python
# 키워드 파라미터 **가 붙은건 키워드 파라미터로 인식함
# keyword arguments
# 이것도 역시 관례적으로 kwargs로 정의하여 사용한다.
def test2(**kwargs):
    print(type(kwargs))
    for key, value in kwargs.items():
        print('key:', key, ', value:', value)
    
test2(a = 1, c = 2, b = 3, name ='Bob', age = 29) #유효한 호출
```

## Lambda 함수  
단일문으로 표현되는 익명함수  
코드 상에서 한번만 사용되는 기능이 있을 때, 굳이 함수로 만들지 않고 1회성으로 만들어서 쓸 때 사용.  

```python
square = lambda x:x**2
print(square(5))
```

```python
def add(x, y):
    return x + y

add2 = lambda x, y : x + y
print(add(10, 20))
print(add2(10, 20))
```

```python
strings = ['bob', 'charles', 'alexander3', 'teddy']
strings.sort(key = lambda s : len(s))

print(strings)
```

filter, map, reduce  
-> lambda가 유용하게 사용되는 3가지 대표적 함수 
filter : 특정 조건을 만족하는 요소만 남기고 필터링  
map : 각 원소를 주어진 수식에 따라 변형하여 새로운 리스트를 변환  
reduce : 차례대로 앞 2개의 원소를 가지고 연산. 연사의 결과가 또 다음 연산의 입력으로 진행됨. 마지막까지 진행되면 최종 출력은 1개만 남음

```python
#filter(함수, 리스트) : 리스트가 함수의 조건에 참이 되면 살아남고, 거짓이 되는 요소는 지워짐.

nums = [1, 2, 3, 5, 6, 8, 9]

#짝수만 남기는 filter 예제
list(filter(lambda n : n % 2 == 0, nums))
```

```python
#map
nums = [1, 2, 3, 5, 6, 8, 9, 10, 11, 13, 15]

list(map(lambda n : n**2, nums))
```

```python
list(map(lambda n : n%2 == 0, nums))
```

```python
#reduce
# 다른 모듈에 편입되어 임포트 해야함.
import functools

#functools.reduce(함수, 리스트)

a = list(range(1, 11)) # 1 ~ 10
#리스트 내의 모든 숫자의 합
functools.reduce(lambda x,y: x + y, a)
```

### 함수 연습문제  
##### 1. 주어진 숫자 리스트의 평균을 구하는 함수를 출력하시오.
##### 2. 해당 숫자가 소수인지 아닌지 판별하시오.
##### 3. 2부터 해당 숫자사이에 소수가 몇개인지 출력하는 함수를 구하시오.

```python
# 입력 : 숫자 리스트
# 출력 : 숫자 리스트의 평균값

def mean(nums):
    return sum(nums) / len(nums)

print(mean([1, 2, 3]))
print(mean([1, 2, 3, 4, 5]))
print(mean([1, 2, 3, 4, 5, 7.3, 8.7]))
```

```python
def is_prime(num):
    for n in range(2, num):
        if num % n == 0:
            return False
    return True


print(is_prime(17))
print(is_prime(88))
```

```python
def num_prime(num):
    count = 0
    for n in range(2, num + 1): # num 본인 포함
        if is_prime(n):
            count += 1
    return count

print(num_prime(50))
```
