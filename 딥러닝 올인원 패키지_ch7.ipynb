## 정규표현식(RE, Regular Expression)

```python
# 특정한 패턴과 일치하는 문자열을 '검색', '치환', '제거'하는 기능을 지원
# 이메일 형식 판별, 전화번호 형식 판별 등
```

```python
#raw string : 문자열 앞에 r이 붙으면 해당 문자열이 구성된 그대로 문자열로 반환

a = 'abcdef\n' # escape 문자열
print(a)

#정의한 그대로 표현
b = r'abcdef\n'
print(b)
```

.(마침표) - 어떤 한개의 character와 일치  
\w - 문자 character와 일치 [a-zA-Z0-9]  
\s - 공백문자와 일치  
\t, \n, \r - tab, newline, return  
\d - 숫자 character와 일치[0-9]  
^ - 시작, $ - 끝 각각 문자열의 시작과 끝을 의미
\가 붙으면 스페셜한 의미가 사라진다.(예, \.는 .자체를 의미 \\는 \를 의미)  

### search method  
첫번째로 패턴을 찾으면 match 객체 반환  
패턴을 찾지 못하면 None 반환

```python
import re
m = re.search(r'abc', '123abcdef') # 'abc' 패턴을 가지고 있는가?
print(m.start())
print(m.end())
print(m.group())
```

```python
m = re.search(r'\d\d\d\d', '112adcdef119')
m
```

```python
m = re.search(r'\d\d\d\w', '112adcdef119')
m
```

```python
m = re.search(r'..\w\w', '@#$%ABCDabcd')
m
```
