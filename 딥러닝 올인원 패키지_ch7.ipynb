### [] : 문자들의 범위를 나타내기 위해 사용
- [] 내부의 메타 캐릭터는 캐릭터 자체를 나타낸다.  
- [abck] : a or b or c or k  
- [abc.^] : a or b or c or . or ^  
- [a-d] : -와 함께 사용되면 해당 문자 사이의 범위에 속하는 문자 중 하나  
- [0-9] : 모든 숫자  
- [a-z] : 모든 소문자  
- [A-Z] : 모든 대문자  
- [a-zA-Z0-9] : 모든 앞라벳 문자 및 숫자  
- [^0-9] : ^가 맨 앞에 사용 되는 경우 해당 문자 패턴이 아닌 것과 매칭  

```python
re.search(r'[cbm]at', 'mat')
```

```python
re.search(r'[0-9]haha', 'ghahaha6hahahah')
```

```python
re.search(r'[abc.☆]aron', 'aron☆ia')
```

```python
# ^이 앞에 쓰이면 뒤에 있는것의 부정을 의미 : ^abc -> abc가 아닌것
re.search(r'[^abc]aron','wefaronsdabcger')
```

메타 캐릭터가 캐릭터 자체를 표현하도록 할 경우 사용  
- \., \\

### 반복 패턴  
패턴 **뒤에** 위치하는 *,+,?는 해당 패턴이 반복적으로 존재하는지 검사  
- '+' -> 1번 이상의 패턴이 발생
- '*' -> 0번 이상의 패턴이 발생
- '?' -> 0 혹은 1번의 패턴이 발생
  
반복을 패턴의 경우 greedy하게 검색한다. 가능한 많은 부분이 매칭되도록 한다.

```python
re.search(r'a[bcd]*b', 'abcbdccb') # a로 시작하고 b, c, d 중 하나가 0번이 상 반복하고 b로 끝남.
```

```python
re.search(r'b\w+a','bananananana') #맨앞은 b로 시작하고 중간에 어떠한 문자가 한번이상 반복하고 마지막은 a
```

```python
re.search(r'i+', 'piigiiiiiii') # 맨앞은 i로 시작하는데 1번이상 반복함
```

```python
re.search(r'pi+g', 'pig') #pi로 시작하고 한번이상 반복하며 g로 끝남
```

```python
re.search(r'https?','httpk://www.naver.com') # http로 시작하고 뒤에 s는 0번 혹은 1번 이상 반복
```

### ^&#42;, &#42;&#36;    
^ : 문자열의 맨 앞부터 일치하는 경우 탐색  
$ : 문자열의 맨 뒤부터 일치하는 경우 검색

```python
# b로 시작하고 중간에 어떠한 단어가 1번 이상 반복하고 a로 끝 -> bana
re.search(r'b\w+a', 'cabana')
```

```python
# 이전 조건에서 맨앞부터 탐색을 실시하는 것이 추가됨
# 실행 결과는 x
re.search(r'^b\w+a', 'cabana')
```

```python
# babana
re.search(r'^b\w+a', 'babana')
```

```python
# b로시작하고 어떠한 문자가 한번이상 반복하며 a로 끝남
# (문자열의 맨 뒤부터 일치하는 경우 탐색)-> 끝이 a로 끝나야함
# bana
re.search(r'b\w+a$', 'cabanap')
```

```python
# 검색 안됨 
re.search(r'b\w+a$', 'cabanap')
```

### grouping
()를 사용  
매칭 결과를 각 그룹별로 분리 가능  
패턴 명시할 때, 각 그룹을 괄호() 안에 넣어 분리하여 사용

```python
m = re.search(r'(\w+)@(.+)', 'test@gmail.com')
print(m.group(0))
print(m.group(1))
print(m.group(2))
```

### {}
- *,+,?을 사용하여 반복적인 패턴을 찾는 것이 가능하나, 반복의 횟수 제한은 불가  
- 패턴 뒤에 위치하는 중괄호 {}에 숫자를 넣으면 해당 숫자 만큼의 경우에만 매칭  
- {4} - 4번 반복, {3,4} - 3~4번 반복

```python
#pi를 3~5번 반복하며 g로 끝남

re.search(r'pi{3,5}g', 'piig')
```

```python
#안나옴
re.search(r'pi{3,5}g', 'piig')
```

```python
#안나옴
re.search(r'pi{3,5}g', 'piiiiiig')
```

### 미니멈 매칭(non-greedy way)
- *?, +?을 이용하여 non-greedy하게 

```python
#<로 시작하고 어떠한 것이 1번이상 반복, >로 끝
#<html>부터 끝까지
re.search(r'<.+>', '<html>haha</html>')
```

```python
re.search(r'<.+?>', '<html>haha</html>')
```

###  {}?
{m,n}? 경우, 최소 m번 매칭하면 만족

```python
#3번만 나옴
re.search(r'a{3, 5}', 'aaaa')
```

## match
- search와 유사하나, 주어진 문자열의 시작부터 비교하여 패턴이 있는지 확인
- 시작부터 없으면 None 반환

```python
re.match(r'\d\d\d', 'my number is 123')
```

```python
re.match(r'\d\d\d', '123 is my number')
```

## findall
- 매칭되는 전체의 패턴을 반환
- 매칭되는 모든 결과를 리스트 형태로 반환

```python
re.findall(r'[\w-]+@[\w.]+', 'test@gmail.com haha test2@gmail.com nice test test')
```

### sub
- 일치하는 모든 패턴을 replace
- 그 결과를 다시 문자열로 반환함
- 두번째 인자는 특정 문자열, 함수가 될 수 있음
- count가  0이면 전체, 1이상이면 해당 숫자만큼 치환

```python
re.sub(r'[\w-]+@[\w.]+', 'great','test@gmail.com haha test2@gmail.com nice test test', 1)
```

### compile
- 정규표현식을 객체로 저장하여 다시 쓸 수 있음

```python
email_reg = re.compile(r'[\w-]+@[\w.]+')

email_reg.search('test@gmail.com haha good')
```

```python
webs = ['https://www.test.co.kr', 
       'https://www.test1.com',
       'http://www.test.com',
       'ftp://www.test.com',
       'http:://www.test.com',
       'htp://www.test.com',
       'http://www.google.com',
       'https://www.homepage.com.']
```

```python
web_reg = re.compile(r'https?://[\w.]+\w+$')
```
